```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://finance.yahoo.com/most-active'
```

```python
response=requests.get(url)
response.text
```

```python
response
```

```python
html_content=response.content
```

```python
soup=BeautifulSoup(html_content,'html.parser')
```

```python
print(soup.prettify)
```

```python
stock_table=soup.find('tbody')
```

```python
codes=[ ]
for listing in stock_table.find_all('tr'):
  sym=listing.find('td',attrs={'aria-label':"Symbol"})
  codes.append(sym.text)
```

```python
codes
```

```python
names=[]
for listing in stock_table.find_all('tr'):
  name=listing.find('td',attrs={'aria-label':"Name"})
  names.append(name.text)
```

```python
names
```

```python
prices=[]
for listing in stock_table.find_all('tr'):
  price=listing.find('td',attrs={'aria-label':"Price (Intraday)"})
  prices.append(price.text)
```

```python
prices
```

```python
aria-label="Change"
```

```python
changes=[]
for listing in stock_table.find_all('tr'):
  change=listing.find('td',attrs={'aria-label':"Change"})
  changes.append(change)
```

```python
changes
```

```python
volumes=[]
for listing in stock_table.find_all('tr'):
  volume=listing.find('td',attrs={'aria-label':"Volume"})
  volumes.append(volume.text)
```

```python
volumes
```

```python
volumes=[]
for listing in stock_table.find_all('tr'):
  volume=listing.find('td',attrs={'aria-label':"Volume"})
  volumes.append(volume.text)
```

```python

per_changes=[]
for listing in stock_table.find_all('tr'):
  per_change=listing.find('td',attrs={'aria-label':"% Change"})
  per_changes.append(per_change.text)
```

```python
import pandas as pd
```

```python
a={
  'Symbol':codes,
   'Name':names,
   'Price':prices,
   'Change':changes,
   '% Change':per_changes,
   'Volume':volumes
  
}
```

```python
df=pd.DataFrame(
    {
          'Symbol':codes,
          'Name':names,
          'Price':prices,
          'Change':changes,
          '% Change':per_changes,
          'Volume':volumes
 
        
    }
)
```
